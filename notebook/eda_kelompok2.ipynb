Notebook ini digunakan untuk menuliskan proses EDA (analisis deskriptif dan visualisasi data).

```python
# Sel 1: Import library dan muat data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys

# Menambahkan folder root ke path agar bisa import dari src
# Ini memungkinkan notebook untuk menemukan modul di folder 'src'
sys.path.append(os.path.abspath(os.path.join('..')))
from src import config

# Memuat data mentah menggunakan path dari config
df = pd.read_csv(config.RAW_DATA_PATH)

print("Data berhasil dimuat. Jumlah baris:", len(df))
df.head()
```

```python
# Sel 2: Informasi dasar dan statistik deskriptif
# --- Gunakan sel Markdown di atas ini dengan judul: "1. Tinjauan Umum Data" ---
print("Informasi Dataset:")
df.info()

print("\nStatistik Deskriptif untuk Kolom Numerik:")
df.describe().T
```

```python
# Sel 3: Pengecekan nilai kosong (missing values)
# --- Gunakan sel Markdown di atas ini dengan judul: "2. Analisis Nilai Kosong" ---
print("Jumlah nilai kosong per kolom:")
df.isnull().sum()
```

```python
# Sel 4: Visualisasi Distribusi Fitur
# --- Gunakan sel Markdown di atas ini dengan judul: "3. Visualisasi Distribusi Fitur" ---

# Fitur yang sudah ada di data mentah
raw_features_for_eda = ['la', 'ld', 'nf', 'ent', 'nd', 'ns', 'aexp', 'arexp', 'asexp']

df[raw_features_for_eda].hist(bins=30, figsize=(15, 10), layout=(3, 3))
plt.suptitle("Histogram Distribusi Fitur Numerik (Data Mentah)", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Sel 5: Visualisasi Heatmap Korelasi
# --- Gunakan sel Markdown di atas ini dengan judul: "4. Analisis Korelasi" ---
plt.figure(figsize=(12, 10))

# Kita hanya gunakan fitur yang sudah ada di data mentah untuk korelasi
correlation_data = df[raw_features_for_eda].copy()
correlation_data['buggy'] = df['buggy'].astype(int) # Tambahkan target untuk melihat korelasinya

sns.heatmap(correlation_data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Heatmap Korelasi Antar Fitur dan Target", fontsize=16)
plt.show()
```
