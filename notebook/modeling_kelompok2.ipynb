Notebook ini digunakan untuk membangun dan mengevaluasi model Machine Learning/Data Mining.

## Langkah yang disarankan:
- Pisahkan data latih dan data uji
- Latih model (misalnya: Decision Tree, Random Forest, SVM, dll.)
- Evaluasi menggunakan metrik: accuracy, precision, recall, F1-score, ROC, confusion matrix

```python
# Sel 1 (BARU): Muat data yang sudah diproses
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os

# Menambahkan folder root ke path
sys.path.append(os.path.abspath(os.path.join('..')))

# Import fungsi dari modul .py kita
from src import config
from src.model import train_and_save_model
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, RocCurveDisplay

# --- Langsung muat data bersih dari folder 'processed' ---
X_train = pd.read_csv(os.path.join(config.PROCESSED_DATA_PATH, "X_train_scaled.csv"))
X_test = pd.read_csv(os.path.join(config.PROCESSED_DATA_PATH, "X_test_scaled.csv"))
y_train = pd.read_csv(os.path.join(config.PROCESSED_DATA_PATH, "y_train.csv")).squeeze('columns')
y_test = pd.read_csv(os.path.join(config.PROCESSED_DATA_PATH, "y_test.csv")).squeeze('columns')


print("Data bersih berhasil dimuat dari 'data/processed/'. Siap untuk pemodelan!")
print(f"Bentuk X_train: {X_train.shape}")
print(f"Bentuk y_train: {y_train.shape}")
```

```python
# Sel 2: Latih dan Evaluasi Model Random Forest
# --- Gunakan sel Markdown di atas ini dengan judul: "1. Model: Random Forest" ---
model_name_rf = "RandomForest"
rf_model = train_and_save_model(X_train, y_train, model_name_rf)

# Tampilkan laporan teks
print("\nLaporan Klasifikasi - Random Forest:")
print(classification_report(y_test, rf_model.predict(X_test)))

# Tampilkan visualisasi
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"Hasil Evaluasi - {model_name_rf}")
ConfusionMatrixDisplay.from_estimator(rf_model, X_test, y_test, ax=ax[0], cmap='Blues')
RocCurveDisplay.from_estimator(rf_model, X_test, y_test, ax=ax[1])
plt.show()
```

```python
# Sel 3: Latih dan Evaluasi Model XGBoost
# --- Gunakan sel Markdown di atas ini dengan judul: "2. Model: XGBoost" ---
model_name_xgb = "XGBoost"
xgb_model = train_and_save_model(X_train, y_train, model_name_xgb)

# Tampilkan laporan teks
print("\nLaporan Klasifikasi - XGBoost:")
print(classification_report(y_test, xgb_model.predict(X_test)))

# Tampilkan visualisasi
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"Hasil Evaluasi - {model_name_xgb}")
ConfusionMatrixDisplay.from_estimator(xgb_model, X_test, y_test, ax=ax[0], cmap='Blues')
RocCurveDisplay.from_estimator(xgb_model, X_test, y_test, ax=ax[1])
plt.show()
```

```python
# Sel 4: Latih dan Evaluasi Model Logistic Regression
# --- Gunakan sel Markdown di atas ini dengan judul: "3. Model: Logistic Regression" ---
model_name_lr = "LogisticRegression"
lr_model = train_and_save_model(X_train, y_train, model_name_lr)

# Tampilkan laporan teks
print("\nLaporan Klasifikasi - Logistic Regression:")
print(classification_report(y_test, lr_model.predict(X_test)))

# Tampilkan visualisasi
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"Hasil Evaluasi - {model_name_lr}")
ConfusionMatrixDisplay.from_estimator(lr_model, X_test, y_test, ax=ax[0], cmap='Blues')
RocCurveDisplay.from_estimator(lr_model, X_test, y_test, ax=ax[1])
plt.show()
```

```python
# Sel 5: Latih dan Evaluasi Model K-Nearest Neighbors (KNN)
# --- Gunakan sel Markdown di atas ini dengan judul: "4. Model: K-Nearest Neighbors" ---
model_name_knn = "KNeighbors"
knn_model = train_and_save_model(X_train, y_train, model_name_knn)

# Tampilkan laporan teks
print("\nLaporan Klasifikasi - K-Nearest Neighbors:")
print(classification_report(y_test, knn_model.predict(X_test)))

# Tampilkan visualisasi
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"Hasil Evaluasi - {model_name_knn}")
ConfusionMatrixDisplay.from_estimator(knn_model, X_test, y_test, ax=ax[0], cmap='Blues')
RocCurveDisplay.from_estimator(knn_model, X_test, y_test, ax=ax[1])
plt.show()
```

```python
# Sel 6: Latih dan Evaluasi Model LightGBM
# --- Gunakan sel Markdown di atas ini dengan judul: "5. Model: LightGBM" ---
model_name_lgbm = "LightGBM"
lgbm_model = train_and_save_model(X_train, y_train, model_name_lgbm)

# Tampilkan laporan teks
print("\nLaporan Klasifikasi - LightGBM:")
print(classification_report(y_test, lgbm_model.predict(X_test)))

# Tampilkan visualisasi
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle(f"Hasil Evaluasi - {model_name_lgbm}")
ConfusionMatrixDisplay.from_estimator(lgbm_model, X_test, y_test, ax=ax[0], cmap='Blues')
RocCurveDisplay.from_estimator(lgbm_model, X_test, y_test, ax=ax[1])
plt.show()
```

```python
# Sel Final: Perbandingan Kurva ROC Semua Model
# --- Gunakan sel Markdown di atas ini dengan judul: "6. Perbandingan Kurva ROC" ---

# Pastikan semua variabel model dari sel-sel sebelumnya sudah ada:
# rf_model, xgb_model, lr_model, knn_model, lgbm_model

# 1. Siapkan kanvas plot
fig, ax = plt.subplots(figsize=(12, 9))

# 2. Plot garis 'No Skill' sebagai acuan
ax.plot([0, 1], [0, 1], 'r--', label='No Skill')

# 3. Buat dictionary untuk memudahkan proses plot
all_models = {
    "Random Forest": rf_model,
    "XGBoost": xgb_model,
    "Logistic Regression": lr_model,
    "K-Nearest Neighbors": knn_model,
    "LightGBM": lgbm_model
}

# 4. Loop untuk memplot setiap kurva model ke kanvas yang sama
for name, model in all_models.items():
    RocCurveDisplay.from_estimator(model, X_test, y_test, name=name, ax=ax)

# 5. Atur judul, legenda, dan tampilkan plot
plt.title('Perbandingan ROC Curve dari Lima Model', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```
